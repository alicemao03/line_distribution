```python
import re
import requests
from datetime import datetime
from synced_lyrics import *
```

```python
def get_meta_data(song, artist, album):
    url = f"https://lrclib.net/api/get?artist_name={artist}&track_name={song}&albumName={album}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        return "song not found"
```

```python
def get_synced_lyrics(song):
    synced_lyrics = ''

    file_path = "../official_times/"+ song + ".srt"
    try:
        with open(file_path, 'r') as f:
            synced_lyrics = f.read()
            print("file found")
            return synced_lyrics
        
    except FileNotFoundError:
        print('file not found', file_path)
        return FileNotFoundError
```
